```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

print(matplotlib.__version__)
print(pd.__version__)
```

### Dataset

```python
df = pd.read_csv('./2020.csv')
df.head()
```

```python
df.rename(columns ={
    'Country name' : 'Country',
    'Perceptions of corruption' : 'Corruption',
    'Freedom to make life choices' : 'Freedom'
}, inplace = True)
df.columns
```

### Bar Plot

```python
df[:5].plot(x='Country',
           y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
           kind ='bar')

plt.title('contoh Bar Plot')
plt.ylabel('nilai')
plt.show()
```

- `df[:5]` digunakan untuk slicing hanya mengambil 5 baris pertama.
- `plot()`
> Digunakan untuk membuat plotingan berbentuk bar plot. membuatuhkan tiga parameter yaitu :
> - parameter pertama yaitu `x='val_colx'` digunakan untuk memasukan kolom df untuk sumbu x
> - parameter kedua yaitu `y ='val_coly'`digunakan utnuk memasukan klom df untuk sumbu y. dapat menggunakan list untuk memasukan value.
> - parameter ketiga yaitu `kind='bar'` digunakan untuk bentuk plotingannya, `kind='bar'` digunakan untuk membuat plotingan berbentuk bar plot

### Line Plot

```python
df[:5].plot(x='Country',
           y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
           kind ='line')

plt.title('contoh line Plot')
plt.ylabel('nilai')
plt.show()
```

- `df[:5]` digunakan untuk slicing hanya mengambil 5 baris pertama.
- `plot()`
> Digunakan untuk membuat plotingan berbentuk line. membuatuhkan tiga parameter yaitu :
> - parameter pertama yaitu `x='val_colx'` digunakan untuk memasukan kolom df untuk sumbu x
> - parameter kedua yaitu `y ='val_coly'`digunakan utnuk memasukan klom df untuk sumbu y. dapat menggunakan list untuk memasukan valuenya
> - parameter ketiga yaitu `kind='line'` digunakan untuk bentuk plotingannya, `kind='line'` digunakan untuk membuat plotingan berbentuk line

```python
df.plot(y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
        kind ='box')

plt.title('contoh line Plot')
plt.ylabel('nilai')
plt.show()
```

- df.plot()

> - parameter pertama digunakan untuk memasukan value sumbu y, valuesnya dapat menggunakan list.

### Scatter plot

```python
df.plot(x='Corruption', y='Freedom',kind='scatter')

plt.xlim((0,1))
plt.ylim((0,1))

plt.title('Contoh Scatter Plot')
plt.show()
```

`df.plot()`
> Digunakan untuk membuat plotingan berbentuk scatter plot. membuatuhkan tiga parameter yaitu :

> - parameter pertama yaitu `x='val_colx'` digunakan untuk memasukan kolom df untuk sumbu x
> - parameter kedua yaitu `y ='val_coly'`digunakan utnuk memasukan klom df untuk sumbu y
> - parameter ketiga yaitu `kind='scatter'` digunakan untuk bentuk plotingannya, `kind='scatter'` digunakan untuk membuat plotingan berbentuk scatter plot.

- `plt.xlim((val_max,val_min))` digunakan untuk memenutkan nilai terbesar dan terkecil dari sumbu x 
- `plt.ylim((val_max,val_min))` digunakan untuk memenutkan nilai terbesar dan terkecil dari sumbu y

```python
df[:5].plot(x='Corruption', y='Freedom',kind='scatter')

plt.xlim((0,1))
plt.ylim((0,1))

plt.title('Contoh Scatter Plot')
plt.show()
```

### Histogram

```python
df.plot(x='Corruption', y='Freedom',kind='hist', bins=10)

plt.title('contoh histogram Plot')
plt.ylabel('nilai')
plt.show()
```

`df.plot()`
> Digunakan untuk membuat plotingan berbentuk histogram plot. membuatuhkan empat parameter yaitu :

> - parameter pertama yaitu `x='val_colx'` digunakan untuk memasukan kolom df untuk sumbu x
> - parameter kedua yaitu `y ='val_coly'`digunakan utnuk memasukan klom df untuk sumbu y
> - parameter ketiga yaitu `kind='hist'` digunakan untuk bentuk plotingannya, `kind='hist'` digunakan untuk membuat plotingan berbentuk histogram plot.
> - parameter keempat yaitu `bins=val` diguankan untuk jumlah class yang akan dibentuk dari data yang dimiliki.


```python
df.plot(x='Corruption', y='Freedom',kind='hist', bins=20)

plt.title('contoh histogram Plot')
plt.ylabel('nilai')
plt.show()
```

bila binsnya di tambahkan maka akan semakin detail

```python
df.plot(x='Corruption', y='Freedom',kind='hist', bins=5)

plt.title('contoh histogram Plot')
plt.ylabel('nilai')
plt.show()
```

bila dikecilkan kelompok datanyapun akan sedikit.

```python
df.plot(x='Country',
        y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
        kind ='hist',
       subplots= True,
       layout = (2,2))

plt.suptitle('contoh subplots')
plt.tight_layout()
plt.show()
```

- menampilkan data dari data yang berbeda.
- `df.plot()`
> Digunakan untuk membuat plotingan berbentuk histogram plot. membuatuhkan empat parameter yaitu :

> - parameter pertama yaitu `x='val_colx'` digunakan untuk memasukan kolom df untuk sumbu x
> - parameter kedua yaitu `y ='val_coly'`digunakan utnuk memasukan klom df untuk sumbu y
> - parameter ketiga yaitu `kind='hist'` digunakan untuk bentuk plotingannya, `kind='hist'` digunakan untuk membuat plotingan berbentuk histogram plot.
> - parameter keempat yaitu `subplots = True` digunakan untuk membuat data yang dapat di akses yang berbeda digunakan subplots.
> - parameter kelima yaitu `layout = (val_rows,val_cols)` digunakan untuk mengatur layout 2 baris dan 2 kolom
- `plt.suptitle()` digunakan untuk menaruh title yang lebih tinggi, suptitle juga di sebut 'super title'
- `plt.tight_layout()` digunakan untuk menghindari overlapping.

- ### Mengenal Twin Axes

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Kasus 1 Twin Axes

```python
x = np.linspace(1,10,25)
x
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x,np.exp(x),'bs-',label ='exp')
ax1.set_xlabel('sumbu X')
ax1.set_ylabel('exp')

ax2 = ax1.twinx()
ax2.plot(x, np.log(x), 'ro-', label ='log')
ax2.set_ylabel('log')

fig.suptitle('contoh twin axes')
fig.legend(loc='upper left')

plt.show()
```

- pembuatan twin axes ini menggunakan **OO Style.**
- twin axes ini terdapat **satu sumbu x dan dua sumbu y.** 
- sumbu y pertama yaitu **exponensial** (berwarna biru pada figure diatas). 
- sumbu y kedua yaitu **logaritme** (berwarna merah pada figure diatas).
- penggunaan parameter tiga pada `plot()` digunakan untuk **mengatur warna, marker, dan garis** (`bs-`) bisa diartikan b yaitu blue, s artinya square, dan - yaitu garis.
- `label = 'value_string'` pada plot() digunakan untuk **memberikan nama pada plotnya**, akan terlihat pada fungsi `legend()`.
- penggunaan `loc = 'value_loc'`pada `legend()` digunakan untuk **penepatan lokasi legend.**
- untuk melakukan **title pada figure** menggunakan `suptitle()`.

### Kasus 2 Twin Axes

```python
x = np.random.randint(16,40, size=30)
x
```

```python
def C2F(celsius = 0):
    return (celsius * 1.8) + 32

def konversi_sumbu(ax1):
    y1,y2 = ax1.get_ylim()
    ax2.set_ylim(C2F(y1), C2F(y2))
    ax2.figure.canvas.draw()
    
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.callbacks.connect('ylim_changed', konversi_sumbu)
ax1.plot(x)

ax1.set_xlabel('hari')
ax1.set_ylabel('Celsius')
ax2.set_ylabel('Fahrenheit')

fig.suptitle('Tempartur udara')
plt.show()
```

- penggunaan `get_ylim()` yaitu digunakan untuk melihat **limit dari y.**
- penggunaan `callbacks.connect()` digunakan untuk **menghubungkan sumbu y dari axes pertama dan axes kedua.** 
- penggunaan `twinx()` digunakan untuk **membuat axes kedua.**

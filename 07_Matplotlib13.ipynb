### 3D Axes

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
plt.show()
```

- cara membuat 3D Axes
> 1. membangun figurenya dulu dengan cara `figure()`
> 2. membentuk project axesnya `axes(projection='3d')`
> 3. `projection='3d'` digunakann untuk membuat project 3Dnya

### 3D Line Plot

```python
z = np.linspace(0,1,100)
x = z* np.sin(20*z)
y = z* np.cos(20*z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x,y,z,color='red')

ax.set(title='contoh 3D Line Plot',
      xlabel = 'sumbu X',
      ylabel = 'sumbu Y',
      zlabel = 'sumbu z')

plt.show()
```

### 3D Scatter Plot

```python
z = np.linspace(0,1,100)
x = z* np.sin(20*z)
y = z* np.cos(20*z)
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.scatter(x,y,z,color ='green')
ax.set(title='Contoh 3D Scatter Plot',
      xlabel='Sumbu X',
      ylabel='Sumbu Y',
      zlabel='Sumbu Z')

plt.show()
```

- cara membuat 3D Scatter Plot
> 1. membangun figurenya dulu dengan cara `figure()`
> 2. membentuk project axesnya `axes(projection='3d')`
> 3. `projection='3d'` digunakann untuk membuat project 3Dnya
> 4. megguakan `scatter(var_x,var_y,var_z,color='val_color')` untuk membuat 3D scatter plot, penggunaann fungsi `scatter()` digunakan untuk membuat sebuah plot dalam bentuk dot atau titik.
>> - parameter kesatu digunakan untuk memberikan value sumbu x
>> - parameter kedua digunakan untuk memberikan value sumbu y
>> - parameter ketiga digunakan untuk memberikan value sumbu z
>> - parameter keempat digunakan untuk memberikan warna pada plotnya.

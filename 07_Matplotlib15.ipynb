```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### dataset

```python
n = np.linspace(-3.0,3.0,100)
x,y = np.meshgrid(n,n)
z = np.sqrt(x**2 + y**2)
print(x.shape)
print(y.shape)
print(z.shape)
```

- `linspace(val_min,val_max,val_sumdata)`
> - parameter kesatu digunakan untuk memberikan value minimum
> - parameter kedua digunakan untuk meberikan value maksimum
> - parameter ketiga digunakan untuk menentukan banyak data.
- `meshgrid((linspace()),(linspace()))`
> diguanakan untuk menghasilkan dua array dan setiap variablenya akan menghasilkan data isi dari `meshgrid()`

### 3D Wireframe

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(x,y,z,color='orange')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_ylabel('z')
ax.set_title('3D Wireframe')

plt.show()
```

- cara membuat 3D WireFrame
> 1. membangun figurenya dulu dengan cara `figure()`
> 2. membentuk project axesnya `axes(projection='3d')`
> 3. `projection='3d'` digunakann untuk membuat project 3Dnya
> 4. lalu menggunakan `plot_wireframe(var_x,var_y,var_z,color='val_color')` untuk membuat ploting wireframe
>> - parameter kesatu digunakan untuk memberikan value sumbu x
>> - parameter kedua digunakan untuk memberikan value sumbu y
>> - parameter ketiga digunakan untuk memberikan value sumbu z
>> - parameter keempat digunakan untuk memberikan warna pada plotnya.


### 3D Surface Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x,y,z, color='orange')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface Plot')

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x,y,z,cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface Plot')

plt.show()
```

- cara membuat surface plot
> 1. membangun figurenya dulu dengan cara `figure()`
> 2. membentuk project axesnya `axes(projection='3d')`
> 3. `projection='3d'` digunakann untuk membuat project 3Dnya
> 4. menggunakan fungsi `plot_surface(var_x,var_y,var_z,color='val_color')`
>> parameter 
>> - parameter kesatu digunakan untuk memberikan value sumbu x
>> - parameter kedua digunakan untuk memberikan value sumbu y
>> - parameter ketiga digunakan untuk memberikan value sumbu z
>> - parameter keempat digunakan untuk memberikan warna pada plotnya.
>> - parameter keempat ini juga dapat menggunakan `cmap=val_cmap` untuk memberikan warna yang berada pada color map (cmap)

### Membuat Pie Plot/ Pie chart

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
```

### Simple Pie Plot

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
```

```python
plt.pie(data, labels = kategori,
        autopct = '%1.1f%%',
        startangle = 90)

plt.title('Contoh Pie Plot')
plt.show()
```

- `var_data` parameter kesatu digunakan untuk memberikan value pada pie plotnya
- `labels = var_labels ` parameter kedua (optional)  digunakan untuk memberikan kategori atau memberi nama.
- `autopct` parameter ketiga (optional) memberitahukan bentuk presentase dari value datanya
- `startangle` untuk melakukan rotasi pada pie plotnya.

### Pengaturan Legend dan Color

```python
kateogri = ['A', 'B','C','D','E']
data = [215, 130, 245, 210, 225]
warna = ['pink','cyan','#e03364', 'yellowgreen','skyblue']
```

```python
plt.pie(data,
       labels = kategori,
       colors = warna,
       autopct = '%1.1f%%',
       startangle = 90)

plt.legend()
plt.title('Contoh Pie Plot')

plt.show()
```

- `colors = var_color` digunakan untuk memberikan warna pada pie plot. warnanya dapat menggunakan kode warna pada paint.
- `plt.legend()` digunakan untuk memberikan informasi tentang pie plotnya.

### Exploded Pie Plot


```python
kategori = ['A','B','C','D','E']
data = [215,130,245,201,225]
warna = ['pink','cyan','#e03364', 'yellowgreen','skyblue']
explode_var = [0.,0.,0.2,0.,0.08]
```

```python
plt.pie(data,
       labels = kategori,
       colors = warna,
       explode = explode_var,
       autopct ='%1.1f%%',
       startangle = 90)

plt.title('Contoh Pie Plot')
plt.show()
```

- `explode = 'var'` digunakan untuk **menarik keluar**. explode diatur dengan nilai **float**, yang berelasi dengan kategori. bila tidak ingin di ledakan maka di beri value `0.`.

### Sample dataset

```python
data = {'Item A' : 109438,
       'Item B' : 103569,
       'Item C' : 112214,
       'Item D' : 112591,
       'Item E' : 100934,
       'Item F' : 103660,
       'Item G' : 137351,
       'Item H' : 123381,
       'Item I' : 135841,
       'Item J' : 104437}

data
```

```python
items = tuple(data.keys())
items
```

```python
count = tuple(data.values())
count
```

> - `tuple(var.keys())`
>> digunakan untuk menampikan kunci dari data tuple 
> - `tuple(var.values())`
>> digunakan untuk menampikan value dari data tuple

### Simple Plot

```python
fig, ax = plt.subplots()
ax.barh(items,count)
plt.show()
```

- `plt.subplots()`
> fungsi ini akan mengembalikan dua object, pertama figure (`fig`) kedua axes (`ax`).
- `barh(var_tupplekeys, var_tupplevalues)`
> fungsi `barh()` ini digunakan untuk membuat plotingan berbentuk histogram horizontal.

### Pengaturan Tick Label

```python
fig, ax = plt.subplots()
ax.barh(items, count)

labels = ax.get_xticklabels()

# set properties
plt.setp(labels,
        rotation = 45,
        horizontalalignment='right')

#plt.setp(labels,
#        rotation = 90,
#        horizontalalignment='left')
        
plt.show()
```

- `ax.get_xticklabels()`
> fungsi ini digunakan untuk mengatur label pada sumbu x 
- `setp(var_ticklables, rotation = 'val_int', horizontalaligment = 'val_[right|left]')`
> param satu digunakan untuk mengpesifikan bagian yang mau di ubah.
> param kedua digunakan utnuk mengatur rotasi tulisan dalam bentuk drajat.
> param ketiga digunakan untuk perataan.

### Pengaturan Format pada Ticker

```python
from matplotlib.ticker import FuncFormatter
```

```python
def ribuan(x, pos):
    return f'{int(x/1000)}K'
```

```python
fig, ax = plt.subplots()
ax.barh(items,count)
formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)
ax.set(title='Contoh Plot',
      xlabel='Jumlah',
      ylabel='Items')
plt.show()
```

- `from matplotlib.ticker import FuncFormatter`
> digunakan untuk mengatur ticker
- `def ribuan()`
> digunakan untuk melakukan konversi
- `FuncFrmatter()`
> digunakan untuk melakukan formatter
- `set_major_formatter()`
> digunakan untuk memasang fortmatternya

### Pegnaturan label pada sumbu (axis) dan judul (title)

```python
fig, ax = plt.subplots()
ax.barh(items, count)
ax.set(title ='Contoh Plot',
      xlabel = 'Jumlah',
      ylabel = 'Items')
plt.show()
```

- `set()`
> - digunakan untuk melakukan pengaturan pada label
> 1. parameter satu yaitu `title='val_string'` digunakan untuk mengatur title
> 2. parameter kedua yaitu `xlabel= 'val_string'` digunakan untuk mengatur label sumbu x
> 3. parameter ketiga yaitu `ylabel= val_string'` digunakan untuk mengatur label sumbu y

### Penambahan Garis(veritcal/horisontal line) pada plot

```python
fig, ax = plt.subplots()
ax.barh(items,count)

ax.axvline(np.mean(count),
           ls = '-',
           color = 'g')

ax.set(title='Contoh Plot',
      xlabel='Jumlah',
      ylabel='Items')

plt.show()
```

- `axvline(np.mean(val))`
> - digunakan untuk membuat garis titk titk merah.
> 1. parameter pertama yaitu `np.mean(val)` digunakan untuk mengatur letak garis, pada kali ini menggunakan nilai tengah dari variable count.
> 2. parameter kedua yaitu `ls='val_obj'` yaitu linesyle digunakan untuk mengatur bentuk garisnya
> 3. parameter ketiga yaitu `color = 'val_color'` digunakan untuk mengatur warna garisnnya

### Menyimpan Hasil Plot ke dalam suatu file

```python
fig.canvas.get_supported_filetypes()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count),
          ls='--',
          color='r')

ax.set(title='Contoh Plot',
      xlabel='jumlah',
      ylabel='Items')

fig.savefig('sales.png',
           transparent = False,
           dpi = 80)
```

- `fig.canvas.get_supported_filetypes()`
> digunakan untuk menunjukan filetype apa saja yang dapat di save
- `fig.savefig()`
> digunakan untuk mengsave gambar yang telah kita ploting.
> 1. parameter pertama yaitu nama file dan file typenya
> 2. parameter kedua yaitu transparent
> 3. parameter ketiga yaitu dpi atau resolusi
> file gambarnya akan di save sesuai dengan directory file yang kita buat.

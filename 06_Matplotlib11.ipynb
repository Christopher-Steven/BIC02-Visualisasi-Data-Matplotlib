- ### Membuat violin plot / violin chart

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Dataset

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

### Simple Violin Plot

```python
plt.violinplot(data)

plt.title('simple violin plot')
plt.xlabel('sambel X')
plt.ylabel('sumbu Y')

plt.show()
```

- `violinplot()` digunakan untuk membuat **violin plot**, garis pada violin plot digunakan untuk menentukan **nilai terbesar dan nilai terkecil.** bentuk shape ini juga disebut **normal distribution.**

### Pengaturan pada violin plot

```python
plt.violinplot(data,
              showextrema =True,
              showmeans=False,
              showmedians=False,
              quantiles=None)

plt.title('violin plot')
plt.xlabel('sumbu X')
plt.ylabel('sumbu Y')

plt.show()
```

- `showextrema = False` pada `violinplot()` digunakan untuk nilai defaultnya `True` digunakan untuk menampilkan garis tengah yang menghasilkan nilai max dan nilai min
- `showmeans = False` pada `violinplot()` digunakan untuk digunakan untuk memunculkan nilai mean akan muncul garis tengah horizontal
- `showmedians = False` pada `violinplot()` digunakan untuk memunculkan nilai median akan muncul garis horizontal pada nilai mediannya.


#### Memenculkan quantiles

```python
plt.violinplot(data,
              showextrema =False,
              showmeans=False,
              showmedians=False,
              quantiles=[0.25,0.5,0.75])

plt.title('violin plot')
plt.xlabel('sumbu X')
plt.ylabel('sumbu Y')

plt.show()
```

- `quantiles = None` pada `violinplot()` digunakan untuk nilai defaultnya yaitu `None`, cara memunculkan dengan menggunakan `[]` setelah tanda `=` bentuk valuenya yaitu persentase (0.0-1), untuk memunculkan quantiles dapat menggunakan `[0.25,0.5,0.75]`, 0.25 digunakan untuk quantiles pertama, 0.5 yaitu quantiles kedua, 0.75 digunakan untuk memunculkan quantiles ketiga.

#### Memunculkan Means

```python
plt.violinplot(data,
              showextrema =False,
              showmeans=True,
              showmedians=False,
              quantiles=None)

plt.title('violin plot')
plt.xlabel('sumbu X')
plt.ylabel('sumbu Y')

plt.show()
```

- `showmeans = False` pada `violinplot()` digunakan untuk digunakan untuk memunculkan nilai mean akan muncul garis tengah horizontal


#### Menentukan median

```python
plt.violinplot(data,
              showextrema =False,
              showmeans=False,
              showmedians=True,
              quantiles=None)

plt.title('violin plot')
plt.xlabel('sumbu X')
plt.ylabel('sumbu Y')

plt.show()
```

- `showmedians = False` pada `violinplot()` digunakan untuk memunculkan nilai median akan muncul garis horizontal pada nilai mediannya.
- median yang terlihat akan terlihat sama seperti nilai mean karena normal distribution

### Horizontal violin Plot

```python
plt.violinplot(data, vert = False)

plt.title('Horizontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

- penggunaan `vert = False` pada `violinplot()` digunakan untuk membuat bentuk violin plot berbentuk horizontal.

### Multiple Violin Plot


```python
np.random.seed(10)
data1 = np.random.normal(100,10,200)
data2 = np.random.normal(80,30,200)
data3 = np.random.normal(90,20,200)
data4 = np.random.normal(70,25,200)

data = [data1, data2, data3, data4]
```

```python
plt.violinplot(data)

plt.title('Multiple Violin plot')
plt.xlabel('sumbu X')
plt.ylabel('sumbu Y')

plt.show()
```

- cara membuat multiple violin plot dapat menggunakan list. lalu memasukan list itu ke dalam fungsi `violinplot()`

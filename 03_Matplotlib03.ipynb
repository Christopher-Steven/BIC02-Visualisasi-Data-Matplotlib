```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
print(matplotlib.__version__)
print(np.__version__)
```

### Simple Line Plot

```python
x = np.arange(0.0,2.0,0.01)
x
```

- `np.arange(parm1,parm2,parm3)` digunakan untuk memunculkan data `array`
- parameter pertama digunakan untuk **start poin** dan bersifat **inklusif atau ikut disertakan**
- parameter kedua digunakan **nilai akhir** dan bersifat **ekslusif yaitu sebagai pembatas dan tidak diikut sertakan**
- param 3 digunakan untuk **interval nilainya** artinya akan ditambahkan terus sampai ke end poin.

```python
s = np.sin(2*np.pi*x)
s
```

- `np.sin()` digunakan untuk melakukan **sinus**
- `np.pi` yaitu **phi**.

```python
flig, ax = plt.subplots() #00 style
ax.plot(x,s)
ax.set(xlabel = 'nilai x',
       ylabel = 'nilai sin(x)',
       title = 'Visualisasi nilai sin')
ax.grid()
plt.show()
```

- `fig, ax = plt.subplots()`menggunakan ***Object Oritented style***.
- `ax.plot(x,s)`, x digunakan sebagai **sumbu x** dan s sebagai **sinus**
- `ax.set` digunakan untuk memberi nama pada plot dengan menggunakan satu command.
- `ax.set` pada di kode ini menggunakan 3 parameter. yaitu `xlabel`, `ylabel` dan `title` semua bertipe **string**
- `ax.grid` digunakan untuk menampilkan garis garis pada plot

```python
plt.plot(x, s, 'cp')
plt.grid()
plt.xlabel ('nilai x')
plt.ylabel ('nilai sin(x)')
plt.title ('Visualisasi nilai sin')
plt.show()
```

### Membuat Subplots

```python
x1 = np.linspace(0.0,5.0,100)
x2 = np.linspace(0.0,2.0,100)
y1 = np.cos(2*np.pi*x1)
y2 = np.cos(2*np.pi*x2)
```

- `np.linspace(parm1, parm2, parm3)` penggunaan linspace berbeda dengan `arrange`, untuk `linspace` nilai **maksimum** dan **minimum** bersifat **inklusif atau akan di sertakan**.
- parameter pertama digunakan untuk nilai **minimum** 
- parameter kedua digunakan untuk nilai **terbesar** 
- parameter ketiga jumlah data poin yang akan **di-generate** atau menunjukan **sejumlah data** parameter ketiga.
- `np.cos`, digunakan untuk menunjukan nilai **cosinus**.

```python
x1
```

#### Multiple Subplots dengan OO Style

```python
fig, (ax1, ax2) = plt.subplots(2,1)
fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, 'ro-')
ax1.set_ylabel('Nilai $cos(x_1)$')

ax2.plot(x2, y2, 'g.-')
ax2.set_ylabel ('Nilai cos (x2)')

ax2.set_xlabel('Nilai x')
```

- `fig` untuk menampung object **figure atau gambar** dan menggunakan
- `plt.subplots(parm 1, parm 2)` untuk parmeter kesatu yaitu menunjukan baris parameter dua menunjukan kolom
- `fig.suptitle('')` digunakan untuk menuliskan title pada figure  bila figure  lebih dari satu 
- `.set_ylabel("")` untuk memberi nama ylabel.
- lateks style, menggunakan karakter `$` untuk membuat nilai 1nya menjadi subscript. contoh `'Nilai $cos(x_1)$'` maka subscriptnya yaitu cost(x1) menjadi $cost(x1)$.
- untuk parameter ke 3 pada `.plot()` digunakan garis `-` digunakan untuk memberikan garis.

### Multiple Subplots dengan pyplot Style

```python
plt.subplot(2, 1, 1)
plt.plot(x1,y1,'ro-')
plt.title('multiple Subplots')
plt.ylabel('nilai $cos(x_1)$')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'g.-')
plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('nilai $x$')

plt.show()
```

- `plt.subplot` untuk parameter ke 3 digunakan untuk memberi tahu yang mana yang mau di-edit atau diubah.

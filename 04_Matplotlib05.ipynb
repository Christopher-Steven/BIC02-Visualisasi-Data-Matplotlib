### Simple Bar Plot

```python
data = [25, 45, 55, 125, 225]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori, data)
plt.grid()
plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title('contoh bar plot')
plt.show()
```

- `bar(parm1,parm2)` parameter 1 digunakan untuk mengatur kategori apa saja yang akan di tampilkan dan paramter 2 digunakan untuk sederet nilai yang akan ditampilkan sesuai dengan urutan kategorinya.



### Pengaturan Grid dan Color

```python
data = [25, 45, 55, 125, 225]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori, data, color = 'red', alpha = 0.25)
plt.grid(linestyle = '--', linewidth = 2, axis ='both', alpha = 0.50)

plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title('contoh bar plot')

plt.show()
```

- `bar(param1, param2, param3, param4)` digunakan untuk membuat plotingan bar plot.
- parameter ketiga yaitu `color =` digunakan untuk melakukan pengaturan warna pada barnya bentuk valuenya color
- parameter keempat yaitu `alpha = ` digunakan untuk mengatur transparansi valuenya yaitu persentase, semakin tinggi semakin pudar semakin tinggi semakin tebal.
- `grid(param1,param2,param3, param4)` digunakan untuk membuat grid pada bar plot.
- parameter kesatu yaitu `linestyle = ` digunakan untuk mengatur garis gridnya.
- parameter kedua yaitu `linewidth = ` digunakan untuk mengatur ketebalan garis gridnya.
- parameter ketiga yaitu `axis = ` digunakan untuk mengatur arah gridnya. untuk `'x'` bentuk arahannya vertikal dan `'y'` bentuk arahannya horizontal. dan `'both'` bentuk arahannya vertikal dan horizontal
- parameter keempat yaitu `alpha = ` digunakan untuk untuk mengatur transparsi dari garis.

### Horisonatal Bar Plot

```python
data = [25, 45, 55, 125, 225]
kategori = ['A','B','C','D','E']
```

```python
plt.barh(kategori, data)
#plt.barh(kategori[::-1], data[::-1])
plt.grid(linestyle = '--', linewidth = 1, axis = 'x', alpha = 0.75)

plt.xlabel('jumlah')
plt.ylabel('kategori')
plt.title('contoh horisontal bar plot')

plt.show()
```

- `barh(param1, param2)` digunakan untuk membuat horisontal bar plot. parameter yang digunakan sama seperti `bar()`. untuk membalik urutan dapat menggunakan `[::-1]`.

### Stacked Bar Plot

```python
data1 = [25,85,75,40,60]
data2 = [40,35,20,55,10]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori,data1,label ='data1')
#plt.bar(kategori,data2, label ='data2')
plt.bar(kategori,data2,label ='data 2', bottom = data1)

plt.legend()

plt.grid(linestyle='--', linewidth =1, axis='y',alpha =0.75)

plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title('contoh stacked bar plot')
plt.show()
```

- stacked bar plot digunakan untuk melakukan ploting pada dua data set yang berbeda. data set kesatu diberi warna biru, dan data set kedua diberikan warna oren.
- `bar(param4)` untuk parameter 4 yaitu `bottom = var` digunakan untuk menumpukan barnya.
- untuk membalik dapat memindahkan parameter ke empat ini.
- bila melakukan penumpukan tampa adanya parameter `bottom` maka akan terjadi penimpaan ploting dilihat dari data ploting yang pertama terlebih dahulu baru yang kedua.

## grouped bar plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategorikategori = ['A','B','C','D','E']

x = np.arange(len(kategori))
x
```

```python
width = 0.35
plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')
plt.xticks(x, kategori)
plt.grid(linestyle = '--', linewidth = 1, axis='y', alpha =0.75)

plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title('contoh grouped bar plot')
plt.legend()
```

- `width` atau lebar dalam bentuk presentase. digunakan untuk memperbesar atau memperkecil barnya.
- `bar(param1)`untuk parameter peertama yaitu `x-width/2` digunakan sebagai nilai tengahnya.
- pada parameter ketiga yaitu `width` digunakan untuk lebar dari bar plotnya
- `xtricks(param1,param2)`digunakan untuk memberikan pengaturan pada sumbu x. 
- bila tidak menggunakan `xtricks()` maka akan melihat indexnya.

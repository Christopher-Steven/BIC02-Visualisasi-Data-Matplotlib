```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Sampel Data

```python
mu, sigma = 100,15
x = mu+sigma*np.random.randn(1000)
x
```

```python
x.shape
```

- variable `mu` digunakan untuk menemukan nilai mean
- variable `sigma` digunakan untuk nilai standard deviation
- `mu+sigma*np.random.randn(1000)` digunakan untuk normal distribution
- nilai `1000` pada `randn`, digunakan untuk memasukan banyak data yang akan digunakan.
- `mu` digunakan supaya sebaran datanya tidak 0
- `shape` digunakan untuk memberi tahu berapa banyak data poinnya.

### Histogram dengan pyplot style

```python
plt.hist(x, bins = 50, facecolor = 'g', alpha = 0.75)

plt.xlabel ('sumbu x')
plt.ylabel('sumbu y')
plt.title('contoh Histogram')

plt.text(45, 500, '$\mu=100,\ \sigma=15$')
plt.grid()

plt.show()
```

- `plt.hist()` mengharuskan adanya dua parameter yang perlu disertakan
- parameter pertama yaitu **data set**nya,
- parameter kedua yaitu `bins`nya. bins digunakan untuk **pengelompokan data**
- `facecolor` digunakan untuk memberikan warna untuk histogramnya
- `alpha` digunakan untuk transparasi, semakin kecil maka semakin pudar. value alpha yaitu persentase.


#### Histogram dengan OO Style

```python
fig, ax = plt.subplots()
ax.hist(x, bins = 50, facecolor = 'b', alpha = 0.25)

ax.set_xlabel('sumbu x')
ax.set_ylabel('sumbu y')
ax.set_title('Contoh Histogram')

ax.text(45, 500, '$\mu=100,\ \sigma=15$')
ax.grid()

plt.show()
```

- Penggunaan `facecolor` dan `alpha` dapat digunakan untuk kombinasi warna.

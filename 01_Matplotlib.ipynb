## mengenal Object Oriented Sytle untuk plotting

membuat figure dan exes, dan memanggil methods dari keduanya

```python
x = np.linspace(0,2,100) 
x
```

np.lispace digunakan untuk menghasilkan deretan bilangan yang di tentukan oleh 3 parameter

Parameter 1 : Start poinnya (0)

Parameter 2 : End poin (2)

parameter 3 : membentuk data poinnya (100)

```python
fig, ax = plt.subplots() #membuat figure dan sebuah axes

#ploting tiga variant data pada axes
ax.plot(x,x, label = 'linear') 
ax.plot(x,x **2, label = 'quadratic')
ax.plot(x, x **3, label = 'cubic')

#memberikan nama pada ploting
ax.set_xlabel('x label') # menyertakan x-label pada axes
ax.set_ylabel('y label') # menyertakan y-label pada axes
ax.set_title('simple plot') # menyertakan title pada axes
ax.legend() #menyertakan legend
```

garis ke 1 warna biru linear itu garis lurus karena xnya sama terus (x,x)

garis ke 2 warna orange kuardan dapetnya dari **2 atau pangkat 2

garis ke 3 warna hijau kubik dapetnya dari **3 atau pangkat 3

bintang bintang melambangkan pangkat, contoh **n (pangkat n)

## menggukanan pyplot sytle untuk melakukan plotting

mengandalkan pyplot untuk membuat figure dan axes, serta fungsi pada pyplot

```python
x = np.linspace(0,2,100)
x
```

```python
#plotting tiga variant data pada axes
plt.plot(x,x, label = 'linear')
plt.plot(x, x**2, label ='quadratic')
plt.plot(x, x**3, label ='cubic')

plt.xlabel('x label') # menyertakan x-label pada axes
plt.ylabel('y label') # menyertakan y-label pada axes
plt.title("simple plot") # menyertakan title pada axes
```

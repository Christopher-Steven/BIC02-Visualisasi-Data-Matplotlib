## Pengaturan Figure Layouts dengan GridSpec

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Simple 2x2 layout dengan subplots

```python
siswa = ['Cecep','Joko', 'Wati', 'Ani']
uas = [68, 83, 96, 72]

fig, axes = plt.subplots(ncols =2 , nrows = 2, constrained_layout=True)

axes[0][0].bar(siswa, uas)
axes[0][1].pie(uas, labels = siswa)
axes[1][0].pie(uas, labels = siswa, explode=[0, 0, 0, 0.2])
axes[1][1].barh(siswa, uas)
plt.show()
```

### `plt.subplots()`
> - parameter pertama digunakan untuk kolom, `ncols = val_int`
> - parameter kedua digunakan untuk baris, `nrows = val_int`
> - parameter ketiga yaitu `constrained_layout=bool` digunakan untuk menentukan tampilan, bila `False` maka tampilannya akan mendempet, bila `True` maka akan memberikan ruang atau renggang
> - var `fig` digunakan untuk menampung satu buah object figure bukan object axes, bila axes digunakan untuk menampung object axes. di dalam figure akan menampung empat buah axes, karena dari `ncols` dan `nrows`nya 
### cara memasukan object axes.
> - menggunakan `axes[val_obj][val_obj]` dilihat dari kiri ke kanan, untuk menentukan object axes dapat dilihat dari urutan indexnya, bila di baris pertama kurung kurawal pertama akan di isi value 0 dan kurung kurawal kedua akan di isi 0 juga, karena kurung kurawal menunjukan kolom, dan kurung kurawal kedua menunjukan pada baris.

## simple 2x2 layout dengan GridSpec

#### Cara pertama

```python
fig = plt.figure(constrained_layout = True)

gs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.bar(siswa, uas)
ax2.pie(uas, labels = siswa)
ax3.pie(uas, labels = siswa, explode =[0,0,0,0.2])
ax4.barh(siswa, uas)

plt.show()
```

- fignya menggunakan `plt.figure` untuk mengatur `(constrained_layout = True)`
- untuk membentuk gridnya menggunakan GridSpec.
- `GridSpec()`
> - parameter pertama digunakan untuk kolom, `ncols = val_int`
> - parameter kedua digunakan untuk baris, `nrows = val_int`
> - parameter ketiga digunakan untuk `figure=fig`. fig digunakan untuk menampung datanya.
- untuk membuat object axesnya maka menggunakan `fig.add_subplot(gs[val_index,val_index])`


#### Cara kedua

```python
fig = plt.figure(constrained_layout = True)

gs = fig.add_gridspec(ncols=2, nrows=2)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.bar(siswa, uas)
ax2.pie(uas, labels = siswa)
ax3.pie(uas, labels = siswa, explode =[0,0,0,0.2])
ax4.barh(siswa, uas)

plt.show()
```

perbedaan dari cara kedua ini yaitu di bagian `fig.add_gridspec()` dengan cara ini tidak perlu memanggil gridspec dengan figure. parameter yang dipanggil hanya dua yaitu `ncols` dan `nrows`

### Rows Span dan Cols Span dengan GridSpec

```python
fig = plt.figure(constrained_layout= True)
gs = fig.add_gridspec(ncols=2, nrows=2)

ax1 = fig.add_subplot(gs[0,:])
## ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.bar(siswa, uas)
ax2.pie(uas, labels = siswa)
ax3.pie(uas, labels= siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

plt.show()
```

melakukan peluasan kolom pada kolom index ke 0.
caranya yaitu pada membuat object axesnya diberikan `(gs[0,:])` pada bagian `fig.add_subplot()`. titik dua atau `:` digunakan untuk membuat black space, jadinya gridspecnya akan digunakan secara keseluruhan.

### Menggunakan teknik Slicing pada GridSpec

```python
fig = plt.figure(constrained_layout = True)
gs = fig.add_gridspec(ncols=3, nrows=3)

ax1 = fig.add_subplot(gs[0,:])
ax1.set_title('gs[0,:]')

ax2 = fig.add_subplot(gs[1,:-1])
ax2.set_title('gs[1,:-1]')
ax2.barh(siswa, uas)

ax3 = fig.add_subplot(gs[1:,-1])
ax3.set_title('gs[1:,-1]')
ax3.bar(siswa, uas)

ax4 = fig.add_subplot(gs[-1,0])
ax4.set_title('gs[-1,0]')

ax5 = fig.add_subplot(gs[-1,-2])
ax5.set_title('gs[-1,-2]')
plt.show()
```
